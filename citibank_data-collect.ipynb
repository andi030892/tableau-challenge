```python
# Import dependencies
import pandas as pd
```

```python
# List of 2018 file names to be combined 
files2018 = ['citibike_data/JC-201801-citibike-tripdata.csv', 'citibike_data/JC-201802-citibike-tripdata.csv', 'citibike_data/JC-201803-citibike-tripdata.csv', 'citibike_data/JC-201804-citibike-tripdata.csv', 'citibike_data/JC-201805-citibike-tripdata.csv', 'citibike_data/JC-201806-citibike-tripdata.csv', 'citibike_data/JC-201807-citibike-tripdata.csv', 'citibike_data/JC-201808-citibike-tripdata.csv', 'citibike_data/JC-201809-citibike-tripdata.csv', 'citibike_data/JC-201810-citibike-tripdata.csv', 'citibike_data/JC-201811-citibike-tripdata.csv', 'citibike_data/JC-201812-citibike-tripdata.csv']

# Create an empty list to store each dataframe
df2018_list = []

# Loop through each file in the list
for file in files2018:
    # Read the file into a dataframe
    df2018 = pd.read_csv(file)

    # Append the dataframe to the list
    df2018_list.append(df2018)

# Concatenate all dataframes in the list into one dataframe
df2018 = pd.concat(df2018_list)

# Drop rows with null values
df2018 = df2018.dropna()

# Show dataframe
df2018.head()
```

```python
# List of 2019 file names to be combined 
files2019 = ['citibike_data/JC-201901-citibike-tripdata.csv', 'citibike_data/JC-201902-citibike-tripdata.csv', 'citibike_data/JC-201903-citibike-tripdata.csv', 'citibike_data/JC-201904-citibike-tripdata.csv', 'citibike_data/JC-201905-citibike-tripdata.csv', 'citibike_data/JC-201906-citibike-tripdata.csv', 'citibike_data/JC-201907-citibike-tripdata.csv', 'citibike_data/JC-201908-citibike-tripdata.csv', 'citibike_data/JC-201909-citibike-tripdata.csv', 'citibike_data/JC-201910-citibike-tripdata.csv', 'citibike_data/JC-201911-citibike-tripdata.csv', 'citibike_data/JC-201912-citibike-tripdata.csv']

# Create an empty list to store each dataframe
df2019_list = []

# Loop through each file in the list
for file in files2019:
    # Read the file into a dataframe
    df2019 = pd.read_csv(file)

    # Append the dataframe to the list
    df2019_list.append(df2019)

# Concatenate all dataframes in the list into one dataframe
df2019 = pd.concat(df2019_list)

# Drop rows with null values
df2019 = df2019.dropna()

# Show dataframe
df2019
```

```python
# List of 2020 file names to be combined 
files2020 = ['citibike_data/JC-202001-citibike-tripdata.csv', 'citibike_data/JC-202002-citibike-tripdata.csv', 'citibike_data/JC-202003-citibike-tripdata.csv', 'citibike_data/JC-202004-citibike-tripdata.csv', 'citibike_data/JC-202005-citibike-tripdata.csv', 'citibike_data/JC-202006-citibike-tripdata.csv', 'citibike_data/JC-202007-citibike-tripdata.csv', 'citibike_data/JC-202008-citibike-tripdata.csv', 'citibike_data/JC-202009-citibike-tripdata.csv', 'citibike_data/JC-202010-citibike-tripdata.csv', 'citibike_data/JC-202011-citibike-tripdata.csv', 'citibike_data/JC-202012-citibike-tripdata.csv']

# Create an empty list to store each dataframe
df2020_list = []

# Loop through each file in the list
for file in files2020:
    # Read the file into a dataframe
    df2020 = pd.read_csv(file)

    # Append the dataframe to the list
    df2020_list.append(df2020)

# Concatenate all dataframes in the list into one dataframe
df2020 = pd.concat(df2020_list)

# Drop rows with null values
df2020 = df2020.dropna()

# Show dataframe
df2020
```

```python
# Concatenate all dataframes for all the years into one file dataframe
combined_df = pd.concat([df2018, df2019, df2020], axis=0)

# Display the results
combined_df
```

```python
# Save the final dataframe to a CSV file
combined_df.to_csv('citibike_data/citibike.csv', index=False)
```
